```python
import pandas as pd
import numpy as np
```

Load Dataset

```python
df = pd.read_csv("f1_monaco_2024_full_dataset.csv")
df.shape
```

```python
df.head()
```

```python
df = df.dropna(subset=["LapTime_s"])
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
time_cols = ["LapTime_s", "Sector1Time_s", "Sector2Time_s", "Sector3Time_s"]

for col in time_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")



df = df.dropna(subset=time_cols)        
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df.to_csv("f1_monaco_2024_cleaned.csv", index=False)

print("Cleaned dataset saved as f1_monaco_2024_cleaned.csv")
```
